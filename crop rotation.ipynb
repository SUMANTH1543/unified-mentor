```python
import seaborn as sns
```

```python
crop_counts = data['Crop'].value_counts()
```

```python
significant_crops = crop_counts[crop_counts > 1000]
```

```python
#Categorical Data Analysis
# Bar chart for crop counts
plt.figure(figsize=(10, 6))
significant_crops.plot(kind='bar')
plt.xlabel('Crop')
plt.ylabel('Count')
plt.title('Crops with Significant Counts')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Scatter plot for area vs production
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Area', y='Production', data=data)
plt.title('Area vs Production')
plt.xlabel('Area')
plt.ylabel('Production')
plt.show()
```
