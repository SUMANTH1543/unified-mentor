```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv("Amazon Sales data.csv")
```

```python
columns_to_convert=["Unit Price","Unit Cost","Total Revenue","Total Cost","Total Profit"]
data[columns_to_convert]=data[columns_to_convert].astype(int)
print(data)
```

```python
data.head(10)
```

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
```

```python
data['month'] = data['Order Date'].dt.month
data['year'] = data['Order Date'].dt.year
```

```python
data.head(10)
```

```python
data['sales'] = data['Unit Price'] * data['Units Sold']
data.head(10)
```

```python
# Month-wise sales trend
month_wise_sales = data.groupby('month')['sales'].sum().reset_index()
print(month_wise_sales)
```

```python
# Year-wise sales trend
year_wise_sales = data.groupby('year')['sales'].sum().reset_index()
print(year_wise_sales)
```

```python
# Yearly month-wise sales trend
year_month_wise_sales = data.groupby(['year', 'month'])['sales'].sum().reset_index()
print(year_month_wise_sales)
```

```python
# Plot month-wise sales trend
plt.figure(figsize=(10, 6))
sns.barplot(x='month', y='sales', data=month_wise_sales)
plt.title('Month-wise Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
# Plot year-wise sales trend
plt.figure(figsize=(10, 5))
sns.lineplot(x='year', y='sales', data=year_wise_sales, marker='o')
plt.title('Year-wise Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
# Plot yearly month-wise sales trend
plt.figure(figsize=(14, 8))
sns.lineplot(x='month', y='sales', hue='year', data=year_month_wise_sales, marker='o')
plt.title('Yearly Month-wise Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(title='Year')
plt.show()
```
